```python
def start_boss_battle():
    weapons = {'bare hands':0,
               'sword':10,
              'axe':15,
              'katana':25,
              'bazooka':50}
    settings = {'hitpoints':{'easy':300,
                             'normal':500,
                             'hard':700},
                'max_damage':{'easy':7,
                             'normal':10,
                             'hard':25}
               }
    difficulty = input('Please set the difficulty of the boss (easy,normal,hard): ')
    weapon = input(f"Now choose a weapon from the below list that you will enter the battle with? \n \
                    {', '.join(weapons.keys())}")
    
    print(f"You've selected {difficulty} and {weapon}")
    
    player_hitpoints = 100
    boss_hitpoints = settings['hitpoints'][difficulty]
    boss_damage = settings['max_damage'][difficulty]
    weapon_damage = weapons[weapon]
                   
    while player_hitpoints > 0 and boss_hitpoints > 0:
        damage = random.randint(1,50) + weapon_damage
        boss_hitpoints = boss_hitpoints - damage
        if boss_hitpoints <= 0:
            print(f'Player hits the boss for {damage} damage. The boss has 0 hitpoints left. Congrats you win!')
            break
        else:
            print(f'Player hits the boss for {damage} damage. The boss has {boss_hitpoints} hitpoints left.')
            print('\n')

        damage = random.randint(1,boss_damage)

        player_hitpoints = player_hitpoints - damage
        if player_hitpoints <= 0:
            print(f'The boss hits the player for {damage} damage. The player has 0 hitpoints left. \
                    Sorry, you have been defeated')
            break
        else:
            print(f'The boss hits the player for {damage} damage. The player has {player_hitpoints} hitpoints left.')
            print('\n')
                   
    
```

```python
start_boss_battle()
```
